ANALISIS MULTIDIMENSIONAL DE VARIABLES

```python
mean_agexcountry=colombianos_df.groupby('Pa√≠s').mean()['Edad (a√±os)']
mean_agexcountry.sort_values(ascending=False).reset_index().iloc[0:10].plot(kind='bar', x='Pa√≠s', y='Edad (a√±os)')
```

```python
mean_cant_persoxcountry=colombianos_df.groupby('Pa√≠s').mean()['Cantidad de personas']
mean_cant_persoxcountry.sort_values(ascending=False).reset_index().iloc[0:10].plot(kind='bar', x='Pa√≠s', y='Cantidad de personas')
```

```python
sum_cant_persoxcountry=colombianos_df.groupby('Pa√≠s').sum()['Cantidad de personas']
sum_cant_persoxcountry.sort_values(ascending=False).reset_index().iloc[0:10].plot(kind='bar', x='Pa√≠s', y='Cantidad de personas')
```
